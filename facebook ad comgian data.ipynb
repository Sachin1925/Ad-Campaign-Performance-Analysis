```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data (1).csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['reporting_start']=pd.to_datetime(df['reporting_start'],format="mixed")
df['reporting_end']=pd.to_datetime(df['reporting_end'],format="mixed")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.dropna(subset=['total_conversion','approved_conversion'])
```

```python
df.isnull().sum()
```

```python
df=df.dropna(subset=['total_conversion', 'approved_conversion'])
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df = df.copy()
df['campaign_id'] = pd.to_numeric(df['campaign_id'], errors='coerce').astype('Int64')
df['fb_campaign_id'] = pd.to_numeric(df['fb_campaign_id'], errors='coerce').astype('Int64')
```

```python
df.info()
```

```python
df[['min_age','max_age']]=df['age'].str.split('-', expand=True)
```

```python
df[['min_age','max_age']]=df[['min_age','max_age']].astype(int)
```

```python
df['mid_age']=(df['min_age']+df['max_age'])//2
```

```python
df.head(2)
```

```python
df.drop(columns=['age','min_age','max_age'], inplace=True)
```

```python
df.head(2)
```

```python
df.describe()
```

```python
df.info()
```

```python
df['impressions']=df['impressions'].astype(int)
```

```python
df.info()
```

```python
df.describe()
```

```python
df['CTR (%)']=(df['clicks']/df['impressions'])*100
df['CPC'] = df['spent']/df['clicks']
df['CPA']=df['spent']/df['total_conversion']
df.head(10)
```

```python
df.info()
```

# Top 5 campaigns by click

```python
top_clicks= df.nlargest(5,'clicks')[['campaign_id','clicks']]
print(top_clicks)
```

# Top 5 campaigns by conversion

```python
top_conversion= df.nlargest(5,'approved_conversion')[['campaign_id','approved_conversion']]
print(top_conversion)    
```

# Correlation between spent click and conversion

```python
correlation=df[['spent','clicks','approved_conversion']].corr()
print(correlation)
```

# Distribution of Impression , click and Conversion

```python
df['impressions'].hist(bins=20)
plt.title("Distribution of Impression")
plt.xlabel("Impression")
plt.ylabel("Frequency")
plt.show()
```

```python
df['clicks'].hist(bins=25, color="green")
plt.title("Distribution of clicks")
plt.xlabel("Clicks")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
df['approved_conversion'].hist(bins=30, color="orange")
plt.title("Distribution of Conversion")
plt.xlabel("Conversion")
plt.ylabel("Frequency")
plt.show()
```

# CTR By Gender

```python
plt.figure(figsize=(4,4))
sns.barplot(data=df,x='gender',y='CTR (%)' ,color="violet")
plt.title("CTR BY gender")
plt.show()
```

# CTR by Age

```python
sns.barplot(data=df,x='mid_age',y='CTR (%)')
plt.title("CTR BY Age")
plt.show()
```

# spent Vs Approved Conversion

```python
sns.scatterplot(data=df,x="spent",y="approved_conversion",hue="gender")
plt.title("Spent vs ApprovedConversion")
plt.show()
```

# Total Conversion by Compaign 

```python
campaign_perf = df.groupby('campaign_id')[['impressions','clicks','spent','approved_conversion']].sum().reset_index()

sns.barplot(data=campaign_perf, x='campaign_id', y='approved_conversion')
plt.title("Total Conversions by Campaign")
plt.show()
```

```python
campaign_perf['ROI'] = campaign_perf['approved_conversion'] / campaign_perf['spent']

sns.barplot(data=campaign_perf, x='campaign_id', y='ROI')
plt.title("ROI by Campaign")
plt.show()
```

```python
sns.heatmap(df[['impressions','clicks','spent','CTR (%)','CPC','CPA','approved_conversion']].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
